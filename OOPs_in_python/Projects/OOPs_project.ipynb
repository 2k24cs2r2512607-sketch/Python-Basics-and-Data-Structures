```python
import nlpcloud
```

```python
 
class NLPApp:
    def __init__(self):
        self.__database={}
        self.__first_menu()
    def __first_menu(self):
        first_input=input("""
Hi! How would u like to proceed?
              1.Not a member? Register
              2.Already a member? Login
              3. Galti se aa gaye? Exit""")
        if first_input=='1':
            self.__register()
        elif first_input=='2':
            self.__login(self)
        else:
            print("register")
    def __register(self):
        name=input("enter name")
        email=input("enter email")
        password=input("Enter password")
        if email in self.__database:
            print("email already exists")
            self.__second_menu()
        else:
            self.__database[email]=[name,password]
            print("registered")
            self.__second_menu()
             
           
    def __second_menu(self):
        second_input=input("""
Hi! How would u like to proceed?
              1.NER
              2.Language Detection
              3. Sentiment Analysis
              4. Logout""")
        if second_input=='1':
            self.__ner()
        elif second_input=='2':
            self.__language_detection()
        elif second_input=='3':
            self.__sentiment_analysis()
        else:
            print("Exit")


    def __login(self):
        email=input("Enter email")
        password=input("Enter password")
        if email in self.__database:
            if self.__database['email'][1]==password:
                print("login Succesfully")
                self.__second_menu()
            else:
                print("Wrong password")
                self.__login()
        else:
            print("not registered email")
            self.__register()
    def __ner(self):
        para=input("Enter paragraph")
        search=input("What would u like to search")
        
        client = nlpcloud.Client("gpt-oss-120b", "bca58f254f9d77af0e9eb7d89d96e00030cdf876", gpu=True)
        response=client.entities(
           para,
            searched_entity=search
        )
        print(response)
    def __sentiment_analysis(self):
        # para=input("Enter paragraph")
        
        client = nlpcloud.Client("gpt-oss-120b", "bca58f254f9d77af0e9eb7d89d96e00030cdf876", gpu=True)
        res=client.sentiment(
            """I hate convoluted products with poor documentation. I see such products all the time. But NLP Cloud is not like that!""",
            target="NLP Cloud"
        )
        print(res)
```
